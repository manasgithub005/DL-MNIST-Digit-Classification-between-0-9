### 5.Model Building

###### 2 step approach

##### 5.1 Build the architecture

```python
from keras.models import Sequential
from keras.layers import Flatten,Dense
```

```python
model=Sequential() #pipeline
model.add(Flatten(input_shape=(28,28)))  #input size 28*28,its a gray image so 1,for color image put 3
model.add(Dense(units=150,activation='relu')) #hidden layer we call it as Dense
model.add(Dense(units=10,activation='softmax')) #its a 10 digit classification problem so take unit=10
model.summary()                                 #for multiclass classification we use softmax activation
```

##### 5.2 Model  complilation

```python
y_train.shape
```

```python
pd.DataFrame(y_train)
```

```python
model.compile( optimizer='Adam', loss='sparse_categorical_crossentropy',metrics='sparse_categorical_accuracy')
```

### 6.Model Training

```python
model_training=model.fit( x=X_train, y=y_train, batch_size=None, epochs=10,validation_data=(X_test,y_test))
```

### 7.Model Testing

```python
model_result=model.evaluate(x=X_test,y=y_test)
print('Model Loss-',round(model_result[0],2))
print('Model Accuracy-',round(model_result[1],2))
```

### 8.Model Prediction

```python
y_pred=model.predict(X_test)
pd.DataFrame(y_pred)
```

```python
y_test[0]
```

```python
plt.imshow(X_test[0],cmap='gray')
```

```python
import numpy as np
np.argmax(a=y_pred[0])
```

### 9.Model Performance

```python
plt.figure(figsize=(10,8))
plt.plot(model_training.history['loss'],label='Training Loss')
plt.plot(model_training.history['val_loss'],label='Validation Loss')
plt.title('Model Performance-Training Loss Vs Validation Loss',size=18)
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show
```

```python
plt.figure(figsize=(10,8))
plt.plot(model_training.history['sparse_categorical_accuracy'],  label='Training Accuracy')
plt.plot(model_training.history['val_sparse_categorical_accuracy'],label='Validation Accuracy')


plt.title('Model Performance-Training Accuracy Vs Validation Accuracy',size=18)
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show
```

## 2.Use DropOut Layer for Optimization

```python
from keras.layers import Dropout
```

```python
model=Sequential() #pipeline
model.add(Flatten(input_shape=(28,28,1)))#input size 28*28,its a gray image so 1,for color image put 3
model.add(Dropout(rate = 0.20))  # 20% will be inactive via dropout 
model.add(Dense(units=150,activation='relu')) #hidden layer we call it as Dense
model.add(Dense(units=10,activation='softmax')) #its a 10 digit classification problem so take unit=10
model.summary() 
```

### Model Compilation

```python
model.compile( optimizer='Adam', loss='sparse_categorical_crossentropy',metrics='sparse_categorical_accuracy')
```

### Model Training

```python
model_training=model.fit( x=X_train_norm, y=y_train, batch_size=None, epochs=10,validation_data=(X_test_norm,y_test))
```

### Model Evaluation

```python
model_result=model.evaluate(x=X_test_norm,y=y_test)
print('Model Loss-',round(model_result[0],2))
print('Model Accuracy-',round(model_result[1],2))
```
